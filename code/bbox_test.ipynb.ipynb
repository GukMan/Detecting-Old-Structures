```python
def polygon_to_bbox(file_name):
  with open(file_name, 'r') as file:
      data = json.load(file)

  Annotations = data['Learning_Data_Info']['Annotations']
  bbox = []

  if 'polygon' not in Annotations[0]:
    for i in Annotations:
      xywh = [i['bbox'][2]/2, i['bbox'][3]/2]
      xywh.extend( i['bbox'][2:] )
      bbox.append( xywh )
    return bbox

  for i in Annotations:
    X = i['polygon'][::2]
    Y = i['polygon'][1::2]
    points = tuple(zip(X, Y))

    left, right = min(X), max(X)
    bottom, top = min(Y), max(Y)
    
    point = []

    xywh = [ (right+left)/2,
            (top+bottom)/2,
            right-left,
            top-bottom ]

    bbox.append( xywh )

    point.append((left,top,right,bottom))
    print(point)

  return bbox, point
```
